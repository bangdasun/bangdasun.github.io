```python
import os
import pandas as pd
```

```python
def markdown_parser(filename):
    with open(filename, encoding='utf-8') as f:
        text_list = f.readlines()
    
    try:
        meta_data, text = ''.join(text_list).split('<!-- more -->')
    except ValueError:
        meta_data, text = ''.join(text_list).split('<!--more-->')
    
    meta_data_list = meta_data.split('\n')
    meta_data_list_prcsd = [meta_data for meta_data in meta_data_list if len(meta_data) > 0]
    title      = meta_data_list_prcsd[1].split(':')[-1].strip()
    datetime   = meta_data_list_prcsd[2].split(':', 1)[-1].strip()
    category   = meta_data_list_prcsd[3].split(':')[-1].strip()
    tags       = meta_data_list_prcsd[4].split(':')[-1].strip()
    short_desc = meta_data_list_prcsd[6]
    
    text = text.replace('\n', '  ')
    
    return dict(title=title, datetime=datetime, category=category, tags=tags, short_desc=short_desc, text=text)
```

```python
blog_md_file_path = '../source/_posts'
blog_filenames = os.listdir(blog_md_file_path)

blog_json = []

for file in blog_filenames:
    print(f'Read {file}')
    output = markdown_parser(os.path.join(blog_md_file_path, file))
    blog_json.append(output)
    
blog_df = pd.DataFrame(blog_json)
```
